```python
#26
import numpy as np
x = pd.Series([1, 2, 3, 4, 5])
y = pd.Series([6, 7, 8, 9, 10])
p1 = np.sum([(a * a) for a in x])
p2 = np.sum([(b * b) for b in y])
p3 = -1 * np.sum([(2 * a*b) for (a, b) in zip(x, y)])
dist = np.sqrt(np.sum(p1 + p2 + p3))
print(dist)
```
